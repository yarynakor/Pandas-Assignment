```python
# Dependencies and Setup
import pandas as pd

# File to Load 
f = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
f_df = pd.read_csv(f)
f_df
```

## Player Count

* Display the total number of players


```python
#display the number of unique players
num_of_players = f_df["SN"].nunique()
player_count=pd.DataFrame([num_of_players], columns = ["Total Players"])
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Find the number of Unique Items, Average Price, Total Purchases, Revenue
unique_purchase_count = f_df["Item Name"].nunique()
ave_price = f_df["Price"].mean()
total_purchase = f_df["Purchase ID"].count()
revenue = f_df["Price"].sum()
summary_df = pd.DataFrame({"# of Unique Items": [unique_purchase_count],
                           "Total Purchases": [total_purchase],
                          "Average Price": [ave_price],
                          "Revenue": [revenue]})
#Format the DataFrame
summary_df["Average Price"] = summary_df["Average Price"].map("${:.2f}".format)
summary_df["Revenue"] = summary_df["Revenue"].map("${:.2f}".format)
summary_df["Total Purchases"] = summary_df["Total Purchases"].map("{:.0f}".format)
summary_df = summary_df[["# of Unique Items", "Total Purchases", "Average Price", "Revenue"]]
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Finding the count and percentage of male, female  and other players
gender_df = f_df.groupby('Gender')
gen_count = gender_df["SN"].nunique()
gen_per = (gen_count/num_of_players)*100
gen_demo = pd.DataFrame({"Gender Count": gen_count, "Gender %": gen_per})
gen_demo["Gender %"] = gen_demo["Gender %"].map("{:.0f}%".format)
gen_demo = gen_demo[["Gender Count", "Gender %"]]
gen_demo
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Creat bins
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
# Create the names for the four bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
f_df['Age Demographics'] = pd.cut(f_df['Age'], age_bins, labels=group_names)
demographic_group =f_df.groupby("Age Demographics")
f_df
```

## Purchasing Analysis (Age)

```python
age_purchases = f_df.groupby("Age Demographics")
age_purchase_count = age_purchases["SN"].count()
age_ave_price = age_purchases["Price"].mean()
age_pur_total = age_purchases["Price"].sum()
by_name_bought = age_pur_total/age_purchase_count
```

```python
age_summary_df = pd.DataFrame({"Purchase Count": age_purchase_count,
                          "Average Purchase Price": age_ave_price,
                          "Total Purchase Value": age_pur_total,
                          "Average Total Purchase Per Person": by_name_bought})
age_summary_df["Average Purchase Price"] = age_summary_df["Average Purchase Price"].map("${:.2f}".format)
age_summary_df["Total Purchase Value"] = age_summary_df["Total Purchase Value"].map("${:.2f}".format)
age_summary_df["Average Total Purchase Per Person"] = age_summary_df["Average Total Purchase Per Person"].map("{:.2f}".format)
age_summary_df = age_summary_df[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Average Total Purchase Per Person"]]
age_summary_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders = f_df.groupby("SN")
item_count = top_spenders["Price"].count()
ave_top = top_spenders["Price"].mean()
total_pv = top_spenders["Price"].sum()
spender_summary_df = pd.DataFrame({"Purchase Count" :item_count,
                            "Average Purchase Price": ave_top,
                            "Total Purchase Value" : total_pv}).round(2)
spender_summary_df=spender_summary_df.sort_values(by ="Total Purchase Value", ascending = False)
spender_summary_df = spender_summary_df.reindex(columns=["Purchase Count","Average Purchase Price","Total Purchase Value"])
spender_summary_df.style.format({"Average Purchase Price":"${:.2f}","Total Purchase Value": "${:.2f}"})
spender_summary_df.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
most_pop = f_df.groupby(["Item ID", "Item Name", "Price"])
pop_item = most_pop["Item Name"].count()
pop_price = most_pop["Price"].sum()
total_pop_val = pop_item*pop_price
pop_items_summary = pd.DataFrame({"Purchase Count": pop_item,
                         "Total Purchase Value": pop_price}).round(2)
pop_items_summary =pop_items_summary.sort_values(by ="Purchase Count", ascending = False)
pop_items_summary["Total Purchase Value"]=pop_items_summary["Total Purchase Value"].map("${:,.2f}".format)
pop_items_summary.head()
```
